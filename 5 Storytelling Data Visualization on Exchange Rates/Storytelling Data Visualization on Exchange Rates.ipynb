# Storytelling Data Visualization on Exchange Rates

---

## 1. Introduction

In this project, we'll analyze a dataset that describes Euro (symbolized by €) daily exchange rates between 1999 and 2021. The dataset was downloaded from [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) on January 2021.

The objective is to **visualise the evolution of the Euro - US dollar exchange rate from 1999 to 2021**.

---

## 2. Open and Read the Data

We will start by opening and reading the dataset using pandas library.

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates
```

Let's also learn some facts about the columns and values in the dataset.

```python
exchange_rates.info()
```

We can see that:

- There are 5699 rows × 41 columns
- There are null values in some columns
- The data type of each column is object or float64.

---

## 3. Clean the Data

**a. Let's rename the `[US dollar ]` and `Period\Unit:` columns to something more intuitive: `US_dollar` and `Time`**

```python
rename_columns = {'[US dollar ]': 'US_dollar',
                  'Period\\Unit:': 'Time'}
exchange_rates.rename(columns = rename_columns, inplace = True)
print('The column names are: ')
print([c for c in exchange_rates.columns])
```

**b. Convert the `Time` column to datetime data type**

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
print(exchange_rates['Time'].dtype)
```

**c. Sort the rows by `Time` in ascending order and reset the index**

```python
exchange_rates.sort_values('Time', inplace = True)
exchange_rates.reset_index(drop = True, inplace = True)
exchange_rates
```

**d. Isolate the `Time` and the `US_dollar` columns for analysis**

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar
```

**e. Inspect `euro_to_dollar` data for erroneous values**

```python
print('Value counts of Euro-USD exchange rate')
print('USD exchange rate: Frequency')
print(euro_to_dollar['US_dollar'].value_counts())
```

The `-` value appears in the `US dollar` column, probably as a placeholder of missing values. We will not consider these rows in the analysis.

```python
rows_to_drop = euro_to_dollar[euro_to_dollar['US_dollar'] == '-'].index
euro_to_dollar.drop(rows_to_drop, inplace = True)

print('Value counts of Euro-USD exchange rate')
print('USD exchange rate: Frequency')
print(euro_to_dollar['US_dollar'].value_counts())
```

**f. Convert the `US_dollar` column to float data type**

```python
euro_to_dollar['US_dollar'] = pd.to_numeric(euro_to_dollar['US_dollar'])
print(euro_to_dollar['US_dollar'].dtype)
```

---

## 4. Analyze the Data

**a. General trend of the Euro to US dollar exchange range**

Now that we're finished cleaning the data, let's generate a line plot to visualize the evolution of the Euro to US dollar exchange rate.

```python
import matplotlib.pyplot as plt

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])

plt.title('Euro to US dollar exchange rate over time')
plt.xlabel('Time')
plt.ylabel('Euro to US dollar rate')

plt.show()
```

We will use the rolling mean to smooth out the daily exchange rate fluctuations and determine the long-term trends. 

Let's compute and plot the 30-day moving average from the daily exchange rates in the data.

```python
# Calculate the moving average with a window of 30 days
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

# Plot moving average
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])

plt.title('30-day rolling mean of Euro to USD exchange rate')
plt.xlabel('Time')
plt.ylabel('Euro to US dollar rate')

plt.show()
```

**b. 2007-2008 financial crisis**

Next, we will look in more detail on how the Euro to US dollar exchange rate changed during the financial crisis.

We will isolate the data from 2006-2009 and 2007-2008 for this visualisation.

```python
# 2006-2009 data
financial_crisis = euro_to_dollar[
    (euro_to_dollar['Time'].dt.year >= 2006) & 
    (euro_to_dollar['Time'].dt.year <= 2009)
].copy()
financial_crisis
```

```python
# 2007-2008 data
financial_crisis_7_8 = euro_to_dollar[
    (euro_to_dollar['Time'].dt.year >= 2007) &
    (euro_to_dollar['Time'].dt.year <= 2008)
].copy()

financial_crisis_7_8
```

Next, let's customise the line plot to:

- Maximize the data-ink ratio
- Create visual patterns using Gestalt principles
- Guide the audience's attention using pre-attentive attributes

```python
# Add FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Plot data from 2006-2009
fig, ax = plt.subplots(figsize = (8, 3))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'],
       linewidth = 1, color = '#A6D785')

# Colour line plot from 2007-2008
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'],
       linewidth = 3, color = '#e23d28')

# Highlight peak area
ax.axvspan(xmin = 13950, xmax = 14140, ymin = 0.09, alpha = 0.3, color='grey')

# Reset and customise labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 13110
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha = 0.5, fontsize = 11)
    x += 365
    
y = 1.193   
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(13020, y, rate, alpha = 0.5, fontsize = 11)
    y += 0.1
    
# Title and subtitle
ax.text(13020, 1.67, 
        "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight = 'bold')
ax.text(13020, 1.63,
        "Euro-USD exchange rates between 2006 and 2010", size = 12)

# Signature 
ax.text(13020, 1.07, ' ' * 118 + 'Source: European Central Bank',
       color = '#f0f0f0', backgroundcolor = '#4d4d4d', size = 10)

# Grid transparency
ax.grid(alpha = 0.5)

plt.show()
```

**c. Comparison between different US presidents**

We will separate the dataset according to the tenures of the last three US presidents:
- George W. Bush (2001-2009)
- Barack Obama (2009-2017)
- Donald Trump (2017-2021)

```python
# 2001-2009 data 
bush = euro_to_dollar[
    (euro_to_dollar['Time'].dt.year >= 2001) & 
    (euro_to_dollar['Time'].dt.year < 2009)
].copy()

# 2009-2017 data 
obama = euro_to_dollar[
    (euro_to_dollar['Time'].dt.year >= 2009) & 
    (euro_to_dollar['Time'].dt.year < 2017)
].copy()

# 2017-2021 data
trump = euro_to_dollar[
    (euro_to_dollar['Time'].dt.year >= 2017) & 
    (euro_to_dollar['Time'].dt.year < 2021)
].copy()
```

We will show comparatively how the Euro to US dollar rate changed under the last three US presidents.

```python
# Add FiveThirtyEight style
style.use('fivethirtyeight')

# Add subplots
plt.figure(figsize=(12, 6))

ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

# Same y axis for all subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(labels = ['1.0', '1.2', '1.4', '1.6'],
                       alpha = 0.3)

# Ax1: plot data from 2001-2009
ax1.plot(bush['Time'], bush['rolling_mean'], color = '#BF5FFF')

ax1.set_xticks(ax1.get_xticks())
ax1.set_xticklabels(labels = ['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], 
                    alpha = 0.3)

ax1.text(12400, 1.92, 'BUSH', size = 18, weight = 'bold', color = '#BF5FFF')
ax1.text(12100, 1.8, '(2001-2009)', weight = 'bold', alpha = 0.3)

# Ax2: plot data from 2009-2017
ax2.plot(obama['Time'], obama['rolling_mean'], color = '#ffa500')

ax2.set_xticks(ax2.get_xticks())
ax2.set_xticklabels(labels = ['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], 
                    alpha = 0.3)

ax2.text(15100, 1.92, 'OBAMA', size = 18, weight = 'bold', color = '#ffa500')
ax2.text(14900, 1.8, '(2009-2017)', weight = 'bold', alpha = 0.3)

# Ax3: plot data from 2017-2021
ax3.plot(trump['Time'], trump['rolling_mean'], color = '#00B2EE')

ax3.set_xticks(ax3.get_xticks())
ax3.set_xticklabels(labels = ['2017', '', '2018', '', '2019', '', '2020', '', '2021'], 
                    alpha = 0.3)

ax3.text(17700, 1.92, 'TRUMP', size = 18, weight = 'bold', color = '#00B2EE')
ax3.text(17600, 1.8, '(2017-2021)', weight = 'bold', alpha = 0.3)

# Ax4: plot from 2001-2021 
ax4.plot(bush['Time'], bush['rolling_mean'], color = '#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'], color = '#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'], color = '#00B2EE')

ax4.set_xticklabels([])

# Title and subtitle
ax1.text(10500, 2.2, 'EURO-USD rate averaged 1.22 under the last three US presidents', fontsize = 20, weight = 'bold')
ax1.text(10500, 2.1, 'EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)',
        fontsize = 12)

# Signature
ax4.text(10600, 0.65, ' ' * 127 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d', size = 14)

plt.show()
```
